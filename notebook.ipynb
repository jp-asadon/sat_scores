```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
df = pd.read_csv("scores.csv")
```

```python
df = df.dropna()
df.info()
```

```python
df = df.rename(columns={
    "School ID" : "school_id",
    "School Name": "school_name",
    "Borough": "borough",
    "Building Code": "building_code",
    "Street Address": "street_address",
    "City": "city",
    "State": "state",
    "Zip Code": "zip_code",
    "Latitude": "latitude",
    "Longitude": "longitude",
    "Phone Number": "phone_number",
    "Start Time": "start_time",
    "End Time": "end_time",
    "Student Enrollment": "student_enrollment",
    "Percent White": "percent_white",
    "Percent Black": "percent_black",
    "Percent Hispanic": "percent_hispanic",
    "Percent Asian": "percent_asian",
    "Average Score (SAT Math)": "math_score",
    "Average Score (SAT Reading)": "reading_score",
    "Average Score (SAT Writing)": "writing_score",
    "Percent Tested" : "percent_tested",
})
df.head()
```

```python
df["percent_white"] = df["percent_white"].astype(str).str.rstrip('%')
df["percent_black"] = df["percent_black"].astype(str).str.rstrip('%')
df["percent_hispanic"] = df["percent_hispanic"].astype(str).str.rstrip('%')
df["percent_asian"] = df["percent_asian"].astype(str).str.rstrip('%')

df["percent_white"] =pd.to_numeric(df["percent_white"])
df["percent_black"] = pd.to_numeric(df["percent_black"])
df["percent_hispanic"] = pd.to_numeric(df["percent_hispanic"])
df["percent_asian"] = pd.to_numeric(df["percent_asian"])

df["count_white"] = df["student_enrollment"] * (df["percent_white"] / 100)
df["count_black"] = df["student_enrollment"] * (df["percent_black"] /100)
df["count_hispanic"] = df["student_enrollment"] * (df["percent_hispanic"] /100)
df["count_asian"] = df["student_enrollment"] * (df["percent_asian"] /100)
df
```

```python
ethnicity = ["count_white", "count_black", "count_hispanic", "count_asian"]

borough_diversity = df.groupby('borough')[ethnicity].sum().round(2).reset_index()
borough_diversity
```

```python
print(df[["start_time", "end_time"]])


df["start_time"] = pd.to_datetime(df["start_time"])
df["end_time"]  = pd.to_datetime(df["end_time"])


df["total_hours"] = ((df["end_time"] - df["start_time"]).dt.total_seconds() / 3600).round(2)
df
```

```python
df["verbal_score"] = df["reading_score"] + df["writing_score"]

df["total_SAT"] = df["verbal_score"] +df["math_score"]
df.head(10)
```

```python
top_boroughs_verbal = df.groupby("borough")["verbal_score"].mean()

top_boroughs_verbal.plot(kind='bar', title='Top Boroughs by Verbal Score')
plt.show()
```

```python
import matplotlib.pyplot as plt

ethnicity = ["count_white", "count_black", "count_hispanic", "count_asian"]
labels = ['White', 'Black', 'Hispanic', 'Asian']

fig, axs = plt.subplots(2, 3, figsize=(12, 8))
axs = axs.flatten()

for i, row in borough_diversity.iterrows():
    sizes = row[ethnicity].values

    axs[i].pie(
        sizes,
        labels=None,  
        autopct='%1.1f%%',
        startangle=90,
        textprops={'fontsize': 9}
    )
    axs[i].set_title(f"{row['borough']}", fontsize=12)

if len(borough_diversity) < len(axs):
    for j in range(len(borough_diversity), len(axs)):
        fig.delaxes(axs[j])

fig.legend(labels, loc='center right', title="Ethnicity", fontsize=10)

plt.suptitle("Racial Distribution per Borough", fontsize=16)
plt.tight_layout()
plt.subplots_adjust(right=0.85, top=0.88)  
plt.show()
```

```python
# plot 1
reading = df.groupby("borough")["reading_score"].mean()
writing = df.groupby("borough")["writing_score"].mean()
math = df.groupby("borough")["math_score"].mean()
total = df.groupby("borough")["total_SAT"].mean()


plt.figure(figsize=(12, 6)) 


plt.subplot(2,2,1)
plt.bar(reading.index, reading.values, color='skyblue')
plt.title("Average Reading Score by Borough")
plt.xticks(rotation=30)

plt.subplot(2,2,2)
plt.bar(writing.index, writing.values, color='lime')
plt.title("Average Writing Score by Borough")
plt.xticks(rotation=30)

plt.subplot(2,2,3)
plt.bar(math.index, math.values, color="salmon")
plt.title("Average Math Score by Borough")
plt.xticks(rotation=30)

plt.subplot(2,2,4)
plt.bar(total.index, total.values, color='gold')
plt.title("Average Total SAT Score by Borough")
plt.xticks(rotation=30)

plt.tight_layout()
plt.show()
```

How much time is ideal students should invest for higher SAT Score?

```python
time_score = df.groupby("total_hours")["total_SAT"].mean().reset_index().sort_values("total_SAT", ascending=False)

plt.scatter(x=time_score["total_hours"], y=time_score["total_SAT"], color="steelblue")

plt.title("Average SAT Score by Total Hours")
plt.xlabel("Total Hours")
plt.ylabel("Average SAT Score")
plt.grid(True)


plt.xticks(np.arange(5, 10.5, 0.5))  

plt.tight_layout()
plt.show()
```

```python


bins = [5, 6, 7, 8, 9, 10]  # 5.00-5.99, 6.00-6.99, ..., 9.00-9.99

labels = ['5.00-5.99', '6.00-6.99', '7.00-7.99', '8.00-8.99', '9.00-9.99']

df['total_hours_cluster'] = pd.cut(df['total_hours'], bins=bins, labels=labels, right=False)

time_score = df.groupby('total_hours_cluster')['total_SAT'].mean().reset_index().sort_values("total_SAT", ascending=False)

print(time_score)
```

Does more time spent in school leads to higher SAT Score?

```python
time_score = time_score.sort_values("total_hours_cluster")

plt.bar(x=time_score["total_hours_cluster"], height=time_score["total_SAT"], color='skyblue')
plt.title("Average SAT Score by Total Hours Cluster")
plt.xlabel("Total Hours Cluster")
plt.ylabel("Average SAT Score")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
city_school = df.groupby("city")["school_id"].count().reset_index().sort_values("school_id", ascending=False)


plt.barh(city_school["city"], city_school["school_id"])
plt.xlabel("Number of Schools")
plt.ylabel("City")
plt.title("Number of Schools per City")
plt.gca().invert_yaxis() 
for index, value in enumerate(city_school["school_id"]):
    plt.text(value + 0.1, index, str(value), va='center', fontsize=7)
    
plt.tight_layout()
plt.show()
```

```python
city_enrollment = df.groupby("city")["student_enrollment"].mean().reset_index().sort_values("student_enrollment", ascending=False)
city_enrollment

plt.bar(city_enrollment["city"], city_enrollment["student_enrollment"])
plt.xlabel("City")
plt.xticks(rotation=90)
plt.ylabel("Average Student Enrollment")
plt.title("Average Student Enrollment by City")
plt.show()
```
